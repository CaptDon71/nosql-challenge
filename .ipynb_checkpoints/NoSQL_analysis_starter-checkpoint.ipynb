## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
pprint(mongo.list_database_names())
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question:

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": {'$gt': 20}}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)

# Display the first document in the results using pprint
from pprint import pprint
first_doc = establishments.find_one(query)
pprint(first_doc)
```

```python
# Convert the result to a Pandas DataFrame
new_df = pd.DataFrame(establishments.find(query))
# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(new_df)}")
# Display the first 10 rows of the DataFrame
new_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
ext_query = {"LocalAuthorityName": {"$regex": "London", "$options": "i"}, "RatingValue": {"$gte": 4}}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(ext_query)
print(count)
# Display the first document in the results using pprint
pprint(establishments.find_one(ext_query))
```

```python
# Convert the result to a Pandas DataFrame
convert_results = establishments.find(ext_query)
convert_df = pd.DataFrame(convert_results)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(convert_df)}")

# Display the first 10 rows of the DataFrame
convert_df.head(10)
```

```python
# Define the query for establishments in London with a RatingValue >= 4
query = {
    "RatingValue": {"$gte": 4},
    "LocalAuthorityName": {"$regex": "London", "$options": "i"} 
}

# Fetch the results from the database
establishments_in_london = list(establishments.find(query))

# Convert the result to a Pandas DataFrame
df_london_establishments = pd.DataFrame(establishments_in_london)

# Display the number of rows in the DataFrame
print(f"There are {df_london_establishments.shape[0]} establishments in London rating equal or greater than 4.")
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Define the location of the "Penang Flavours" restaurant
location_query = {"BusinessName": "Penang Flavours"}

# Retrieve latitude and longitude of "Penang Flavours"
penang_flavours = establishments.find_one(location_query, {"geocode.latitude": 1, "geocode.longitude": 1})

# Convert Decimal128 values to float using .to_decimal()
latitude = float(penang_flavours["geocode"]["latitude"])
longitude = float(penang_flavours["geocode"]["longitude"])

# Define the degree search range
degree_search = 0.01

# Build the query for establishments with RatingValue = 5 and within the search range of latitude/longitude
query = {
    "RatingValue": 5,
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}
sort = "scores.Hygiene"
limit = 5
# Define the fields to return
fields = {'BusinessName': 1, 'scores': 1, 'geocode.latitude': 1, 'geocode.longitude': 1, 'RatingValue': 1}
# Get the count of documents that match the query
count = db.establishments.count_documents(query)
# Print the number of documents contained in the result
print("Number of matching documents:", count)
first_document = db.establishments.find_one(query)

# Display the first document in the results
print("First document in the results:")
pprint(first_document)

# Get the top 5 establishments that match the query and sort by hygiene score (decending)
top_5_establishments = list(db.establishments.find(query, fields).sort(sort, -1).limit(limit))
```

```python
# Convert result to Pandas DataFrame
df_top_5_nearby = pd.DataFrame(top_5_establishments)

# Print the top 5 results
df_top_5_nearby
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0}}
# Group by Local Authority from highest to lowest
group_query = {'$group': {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}}
sort_values = {'$sort': { 'count': -1 }}

# Create pipeline
pipeline = [match_query, group_query, sort_values]

# Query the aggregation pipeline
agg_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in the result:", len(agg_results))

# Print the first 10 results
pprint(agg_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
agg_result_df = pd.DataFrame(agg_results)
# Display the number of rows in the DataFrame
agg_count = len(agg_result_df)
print("Number of rows:", agg_count)

# Display the first 10 rows of the DataFrame
agg_result_df.head(10)
```
