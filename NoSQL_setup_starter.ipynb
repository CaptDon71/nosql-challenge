5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
establishments.update_many({ "geocode.longitude": {"$type": "string"}, "geocode.latitude": {"$type": "string"}}, [{
            "$set": { "geocode.longitude": {"$toDouble": "$geocode.longitude"}, "geocode.latitude": {"$toDouble": "$geocode.latitude"}}}])
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many({"RatingValue": {"$type": "string"}}, [{"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}])
```

```python
# Check that the coordinates and rating value are now numbers
# Query to find documents where longitude, latitude, or RatingValue are not in the expected types
errors = establishments.find(
    {
        "$or": [
            {"geocode.longitude": {"$not": {"$type": "double"}}},
            {"geocode.latitude": {"$not": {"$type": "double"}}},
            {"RatingValue": {"$nin": [0, 1, 2, 3, 4, 5, None]}},
        ]
    },
    {"_id": 1, "geocode.longitude": 1, "geocode.latitude": 1, "RatingValue": 1}
)

# Print the results to identify problematic documents
for error in errors:
    print(error)
```
